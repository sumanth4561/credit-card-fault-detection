```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import gridspec
import sklearn
import scipy
from sklearn.metrics import classification_report,accuracy_score
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM
from pylab import rcParams

```

```python
# loading the data
crcv = pd.read_csv("C:/Users/suman/OneDrive/Attachments/Desktop/creditcard.csv")
crcv.head()
```

```python

val=crcv[crcv['Class']==0]
fraud=crcv[crcv['Class']==1]
print("The number of fraud transactions are",len(fraud))
print("The number of fraud transactions are",len(val))
x=len(fraud)
y=len(val)
outlierfraction=x/y
print("The ratio is ",outlierfraction)
crcv = crcv.sample(frac = 0.1, random_state = 48) 
print(crcv.shape)
print(crcv.describe())
```

EXPLORATORY DATA ANALYSIS

```python
# count_class=pd.value_counts(crcv['Class'],sort=True)
count_class=pd.value_counts(crcv['Class'],sort=True)
count_class.plot(kind='bar',rot=0)
plt.title("Transaction Class Distribution")
plt.xticks(range(2))
plt.xlabel("Class")
plt.ylabel("Frequency")
```

PLOTTING THE CORRELATION MATRIX

```python
corre=crcv.corr()
fig=plt.figure(figsize=(10,9))
sns.heatmap(corre,vmax=.8,square=True)
plt.show()
```

```python
X=crcv.drop(['Class'],axis=1)
Y=crcv['Class']
xdata=X.values
ydata=Y.values
```

TRAINING AND TESTING THE DATA 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(xdata,ydata,test_size = 0.2,random_state = 42)
```

BUILDING RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=50)
rfc.fit(X_train,Y_train)
z=rfc.predict(X_test)
c=rfc.score(X_test,Y_test)
print(c)
```

# GETTING TO KNOW THE ACCURACY

```python
from sklearn.metrics import classification_report, accuracy_score  
from sklearn.metrics import precision_score, recall_score 
from sklearn.metrics import f1_score, matthews_corrcoef 
from sklearn.metrics import confusion_matrix
acc=accuracy_score(Y_test,z)
prec=precision_score(Y_test, z)
rec=recall_score(Y_test, z)
print("The accuracy is",acc)
print("The precision is",prec)
print("The recall score is",rec)
```

# PLOTTING THE CONFUSION MATRIX

```python
 
LABELS = ['Normal', 'Fraud'] 
conf_matrix = confusion_matrix(Y_test, z) 
plt.figure(figsize =(12, 12)) 
sns.heatmap(conf_matrix, xticklabels = LABELS,  
            yticklabels = LABELS, annot = True, fmt ="d"); 
plt.title("Confusion matrix") 
plt.ylabel('True class') 
plt.xlabel('Predicted class') 
plt.show() 
```
